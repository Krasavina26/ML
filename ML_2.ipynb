11. Пусть задания с кратким ответом будут задания типа В. Соответственно всего по экзамену вопросов класса В: В
, ... Вк
 Посчитайте процент выполненных и невыполненных заданий по каждому вопросу класса В.


```python
for i in range(len(data['Задания с кратким ответом'])):
  data.insert(len(data.columns), f'B{i+1}', 0)
data
```

```python
for i, _ in data.iterrows():
  short_ans = list(data['Задания с кратким ответом'][i])
  for j in range(len(short_ans)):
    data.loc[i, f'B{j+1}']=short_ans[j]

data
```

```python
cols = data.loc[:, 'B1':'B12'].columns
for col in cols:
  print(data.groupby([col]).agg({col:'count'})/data.shape[0] * 100)
```

12. Аналогично и с типом С (ответы с развернутым ответом)

```python
data
```

```python
for i in range(len(data['Задания с развёрнутым ответом'])):
  data.insert(len(data.columns), f'C{i+1}', 0)
data
```

```python
for i in range(0, len(data['Задания с развёрнутым ответом']), 3):
  long_ans = list(data['Задания с развёрнутым ответом'][i])
  for j in range(len(long_ans)):
    data.loc[j, f'C{j+1}']=long_ans[j]

data
```

```python
cols = data.loc[:, 'C1':'C12'].columns
for col in cols:
  print(data.groupby([col]).agg({col:'count'})/data.shape[0] * 100)
```
